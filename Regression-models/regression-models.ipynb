```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

### Linear Regression model
   ### To find Thetha we Normal equation 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(lr.score(X_test, y_test))
intercept = lr.intercept_
slope = lr.coef_

final_y = intercept + slope * X
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X, y, color='red', label='Actual')  # Scatter plot of actual data points
plt.plot(X, final_y, color='blue', label='Predicted')  # Line plot for the regression line
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
from  sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
print(f'Mean absolute Error: {mean_absolute_error(y_test, lr.predict(X_test))}')
print(f'Mean Squared Error: {mean_squared_error(y_test, lr.predict(X_test))}')
print(f'R2 Score: {r2_score(y_test, lr.predict(X_test))}')
```

### Normal Equation

```python
x_b = np.c_[np.ones((100,1)),X]
theta_best = np.linalg.inv(np.transpose(x_b).dot(x_b)).dot(np.transpose(x_b)).dot(y)
y_predict = x_b.dot(theta_best)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X, y, color='red', label='Actual')
plt.plot(X, y_predict, color='blue', label='Predicted')
plt.title('Normal Equation')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mse_normal_eq = mean_squared_error(y, y_predict)
print("Mean Squared Error using Normal Equation:", mse_normal_eq)
```
